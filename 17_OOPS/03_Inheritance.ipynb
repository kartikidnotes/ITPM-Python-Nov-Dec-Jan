```python
# Inheritance : Acquiring the property from parent class to child class 

# Parent : Base class 
# Child : Derived class 

# Types :
# 1. Single : 1 parent - 1 child 
# 2. Multiple : many parent -- 1 Child
# 3. Multilevel : one parent -- multiple child 
# 4. Hierarchical : grandparent - parent- child - grandchild --- 
# 5. Hybrid : mix 
```

```python
# Single Inheirtance 

#parent
class Employee:
    def show(self):
        print("Employee Details ")

#child
class Developer(Employee):
    def code(self):
        print("Developer is writing the code !!! ")


d=Developer()
d.code()
d.show()
```

```python
class Employee:
    def get_Emp(self):
        self.name=input("Enter Employee Name :: ")

class Developer(Employee):
    def get_Dev(self):
        self.language=input("Enter Developer's Language :: ")
        print(self.name," codes in ",self.language)


d=Developer()
d.get_Emp()
d.get_Dev()
```

```python
class Person:
    def get_name(self):
        self.name=input("Enter Name :: ")

class Student(Person):
    def get_student(self):
        self.marks=int(input("Enter MArks :: "))
        print(self.name, " has scored ",self.marks)

s1=Student()
s1.get_name()
s1.get_student()

s2=Student()
s2.get_name()
s2.get_student()
```

```python
class Vehicle:
    def get_details(self):
        self.brand=input("Enter Brand :: ")

class Car(Vehicle):
    def car_details(self):
        self.price=int(input("Enter Price :: "))
        print(self.brand, " costs for ",self.price)


c1=Car()
c1.get_details()
c1.car_details()

c2=Car()
c2.get_details()
c2.car_details()
```

```python
class Bank:
    def get_details(self):
        self.bankname=input("Enter Bank Name :: ")

class Account(Bank):
    def get_account(self):
        self.balance=int(input("Enetr Balance :: "))
        print(self.bankname, " Bank has balance of Rs. ",self.balance)

a=Account()
a.get_details()
a.get_account()
```

```python
# Multiple Inheritance 

class Father:
    def fatherskill(self):
        print("He Drives So Well ")

class Mother:
    def motherskill(self):
        print("She Cooks So Well ")

class Child(Father,Mother):
    def childskill(self):
        print("Child Can Study ")

obj=Child()
obj.fatherskill()
obj.motherskill()
obj.childskill()
```

```python
class Teacher:
    def t(self):
        self.subject=input("ENter Subject :: ")

class Writer:
    def w(self):
        self.book=input("Enter Book Name :: ")

class Student(Teacher,Writer):
    def show(self):
        print(self.subject ,"is taught from a book  ",self.book)

s=Student()
s.t()
s.w()
s.show()
```

```python
class Camera:
    def cam(self):
        self.mp=input("Enter MP :: ")

class Brand:
    def ph(self):
        self.brand=input("Enter Brand :: ")

class SmartPhone(Camera,Brand):
    def disp(self):
        print("SmartPhone has camera of ",self.mp," and it is of ",self.brand," brand")


s=SmartPhone()
s.cam()
s.ph()
s.disp()
```

```python
class Song:
    def s(self):
        self.song=input("Enter SOng Name :: ")

class Dancer:
    def d(self):
        self.style=input("Enter Style Of Dance :: ")

class Performer(Song,Dancer):
    def show(self):
        print("Performer is performing on ",self.song, " with dance style ",self.style)

p=Performer()
p.s()
p.d()
p.show()
```

```python
# Mulitilevel
# a=b , b=c , a=c

class Person:
    def p(self):
        self.name=input("Enter Name :: ")

class Employee(Person):
    def e(self):
        self.id=int(input("Enter ID :: "))

class Manager(Employee):
    def m(self):
        print(self.name,self.id)

m=Manager()
m.p()
m.e()
m.m()
```

```python
class Animal:
    def an(self):
        self.type=input("Enter type of animal :: ")

class Cat(Animal):
    def cat(self):
        self.breed=input("Enter Breed :: ")

class Kitten(Cat):
    def kit(self):
        print("Kitten is of type : ",self.type ," and of breed ",self.breed)


k=Kitten()
k.an()
k.cat()
k.kit()
```

```python
class Brand:
    def b(self):
        self.brand=input("Enter Brand NAme :: ")

class RAM:
    def r(self):
        self.ram=input("Enter RAM ")

class GamingLaptop(Brand,RAM):
    def g(self):
        print("Laptop is of brand :: ",self.brand)
        print("Laptop is having RAM of  :: ",self.ram)


gl=GamingLaptop()
gl.b()
gl.r()
gl.g()
```

```python
class Bank:
    def b(self):
        self.name=input("Enter Bank Name :: ")

class Account(Bank):
    def a(self):
        self.acc=int(input("Enter Account Number :: "))

class Savings(Account):
    def sa(self):
        print("Bank name is :: ",self.name)
        print("Account No  :: ",self.acc)


s=Savings()
s.b()
s.a()
s.sa()
```
