````python
# **Encapsulation in Python (Theory)**

# Encapsulation is one of the core concepts of **Object-Oriented Programming (OOP)**. 
# It means **binding data (variables) and methods (functions) together into a single unit (class)**
# and **restricting direct access to some of the object's components**.

# Def : Wrapping up the data into a single unit 
# ---

# ### 🔹 Simple Definition

# Encapsulation = **Data Hiding + Data Protection**

# It prevents accidental modification of data and ensures controlled access.

# ---

# ### 🔹 Why Encapsulation is Important

# * Protects data from unauthorized access
# * Improves code security
# * Makes code easier to maintain
# * Allows controlled access using methods

# ---

# ### 🔹 How Encapsulation Works in Python

# Python uses **access modifiers** (by convention):

# 1. **Public**

#    * Accessible from anywhere
#    * Example: `self.name`

# 2. **Protected (_single underscore)**

#    * Meant for internal use (not strict)
#    * Example: `self._age`

# 3. **Private (__double underscore)**

#    * Cannot be accessed directly outside the class
#    * Example: `self.__salary`

# ---

# ### 🔹 Example (Conceptual)

# ```python
# class Student:
#     def __init__(self, name, marks):
#         self.name = name        # public
#         self.__marks = marks    # private

#     def get_marks(self):        # getter
#         return self.__marks

#     def set_marks(self, m):     # setter
#         if m >= 0:
#             self.__marks = m
# ```

# ---

# ### 🔹 Key Points

# * Private variables are accessed using **getter and setter methods**
# * Python does not strictly enforce access modifiers, but follows naming conventions
# * Private variables use **name mangling** (`_ClassName__variable`)

# ---

# ### 🔹 Advantages

# * Data security
# * Better control over data
# * Flexibility to change implementation without affecting code

# ---

# ### 🔹 Real-Life Example

# Think of a **bank account**:

# * Balance is hidden (private)
# * You can access it only through methods like `deposit()` or `withdraw()`
````

````python
# public 
class Student:
    def __init__(self):
        self.name="Raj"


s=Student()
print(s.name)
````

````python
# private : __

class Student:
    def __init__(self):
        self.__marks=90

    def get_marks(self):
        return self.__marks
    

s=Student()
# print(s.__marks) # error

print(s.get_marks()) # call -- getter method 
````

````python
# protected : _

class Employee:
    def __init__(self):
        self._salary=60000

class Child(Employee):
    def show(self):
        print(self._salary)

c=Child()
c.show()
````

````python
class Bank:
    def __init__(self):
        self.__balance=0

    def deposit(self,amount):
        self.__balance+=amount

    def get_balance(self):
        return self.__balance
    

b=Bank()
b.deposit(50000)
print(b.get_balance())
````

````python
class Employee:
    def __init__(self):
        self.__salary=0

    def set_salary(self,s):
        if s>0:
            self.__salary=s
    
    def get_salary(self):
        return self.__salary
    
e=Employee()
e.set_salary(99990)
print(e.get_salary())
````

````python
class Cart:
    def __init__(self):
        self.__items=[]

    def add_item(self,item):
        self.__items.append(item)

    def remove_item(self,item):
        if item in self.__items:
            self.__items.remove(item)
    
    def view_cart(self):
        return self.__items
    

c=Cart()
c.add_item("Laptop")
c.add_item("Mouse")
c.add_item("Charger")
print(c.view_cart())

c.remove_item("Mouse")
print(c.view_cart())

````

````python
class User:
    def __init__(self,username,password):
        self.__username=username
        self.__password=password

    def login(self,pwd):
        return pwd==self.__password
    

u=User("admin","1234")
print(u.login("1234"))
print(u.login("Wrong Password"))
````

````python
class ATM:
    def __init__(self,pin,balance):
        self.__pin=pin
        self.__balance=balance

    def check_balance(self,entered_pin):
        if entered_pin==self.__pin:
            print("Balance is :: ",self.__balance)
        else:
            print("Invalid Pin")

atm=ATM(1234,50000)
atm.check_balance(int(input("Enter Pin :: ")))
        
````

````python
class Ride:
    def __init__(self):
        self.__fare=0

    def calculate_fare(self,km):
        self.__fare=km*12

    def get_fare(self):
        print("Total Fare is :: Rs.  ",self.__fare)


r=Ride()
r.calculate_fare(int(input("Enter Distance in km :: ")))
r.get_fare()
````

````python
class Product:
    def __init__(self,price):
        self.__price=price

    def applydiscount(self,percent):
        self.__price-=(self.__price*percent/100)

    def finalprice(self):
        print("Total Final Price :: ",self.__price)

p=Product(10000)
p.applydiscount(int(input("Enter Discount in % :: ")))
p.finalprice()
````

````python
class Loan:
    def __init__(self):
        self.__creditscore=0
        
    def set_score(self,score):
        self.__creditscore=score

    def check_approval(self):
        if self.__creditscore>700:
            print("Loan Approved ")
        else:
            print("Loan Rejected ")

l=Loan()
l.set_score(int(input("Enter Credit Score :: ")))
l.check_approval()
````

````python
class Employee:
    def __init__(self):
        self.__rating=0

    def evaluate(self,work_hours):
        if work_hours>8:
            self.__rating=5
        else:
            self.__rating=3

    def show_rating(self):
        print("Rating is :: ",self.__rating)


e=Employee()
e.evaluate(int(input("Enter Work in Houurs :: ")))
e.show_rating()
````

````python
class Subscription:
    def __init__(self):
        self.__active=False

    def activate(self):
        self.__active=True

    def access_content(self):
        if self.__active:
            print("Access Granted")
        else:
            print("Access denied ")

s=Subscription()
s.access_content()
s.activate()
s.access_content()
````
