```python
# Inheritance : Acquiring the property from parent class to child class 

# Parent : Base class 
# Child : Derived class 

# Types :
# 1. Single : 1 parent - 1 child 
# 2. Multiple : many parent -- 1 Child
# 3. Multilevel : one parent -- multiple child 
# 4. Hierarchical : grandparent - parent- child - grandchild --- 
# 5. Hybrid : mix 
```

```python
# Single Inheirtance 

#parent
class Employee:
    def show(self):
        print("Employee Details ")

#child
class Developer(Employee):
    def code(self):
        print("Developer is writing the code !!! ")


d=Developer()
d.code()
d.show()
```

```python
import math

class Shape:
    def sh(self):
        self.sides = int(input("Enter number of sides: "))


class Rectangle(Shape):
    def rect(self):
        self.length = float(input("Enter length: "))
        self.breadth = float(input("Enter breadth: "))
        area = self.length * self.breadth
        print("Rectangle area =", area)


class Triangle(Shape):
    def tri(self):
        self.base = float(input("Enter base: "))
        self.height = float(input("Enter height: "))
        area = 0.5 * self.base * self.height
        print("Triangle area =", area)


class Circle(Shape):
    def cir(self):
        self.radius = float(input("Enter radius: "))
        area = math.pi * self.radius * self.radius
        print("Circle area =", area)


class Pentagon(Shape):
    def pen(self):
        self.side = float(input("Enter side length: "))
        area = (5 * self.side * self.side) / (4 * math.tan(math.pi/5))
        print("Pentagon area =", area)


# Objects
r = Rectangle()
r.sh()
r.rect()

t = Triangle()
t.sh()
t.tri()

c = Circle()
c.sh()
c.cir()

p = Pentagon()
p.sh()
p.pen()
```

```python
# Hybrid 

class Person:
    def p(self):
        self.name=input("Enter Name :: ")

class Employee(Person):
    def e(self):
        self.id=input("Enter ID :: ")
    
class Student(Person):
    def s(self):
        self.course=input("Course :: ")

class Intern(Employee,Student):
    def stud(self):
        print(self.name,self.id,self.course)

obj=Intern()
obj.p()
obj.e()
obj.s()
obj.stud()
```

```python
class Company:
    def c(self):
        self.name=input("Enter Company Name :: ")

class HR(Company):
    def hr(self):
        self.hires=input("Hires :: ")
    
class Tech(Company):
    def tech(self):
        self.tech=input("Tech :: ")

class Manager(HR,Tech):
    def man(self):
        print("Company Name :: ",self.name)
        print("Hiring Count is  :: ",self.hires)
        print("Technology Name :: ",self.tech)

obj=Manager()
obj.c()
obj.hr()
obj.tech()
obj.man()
```
