```python
class Employee:
    def accept(self,name,salary):
        self.name=name
        self.salary=salary 

    def display(self):
        print("Name is :: ",self.name)
        print("Salary is :: ",self.salary)


e1=Employee()
e2=Employee()


e1.accept("Ram",50000)
e2.accept("Raj",30000)

e1.display()
e2.display()
```

```python
class Employee:
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal

    def display(self):
        print("Name is :: ",self.name)
        print("Salary is :: ",self.sal)

e1=Employee("Ram",50000)
e2=Employee("Raj",30000)

e1.display()
e2.display()
    
    
```

```python
class Bank:
    def __init__(self,balance):
        self.balance=balance

    def display_bal(self):
        print("Current Balance is :: ",self.balance)


b=Bank(50000)
b.display_bal()
        
```

```python
class EvenOdd:
    def check(self,num):
        if num%2==0:
            print("Even ")
        else:
            print("Odd")

n=EvenOdd()
n.check(8)
```

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price


p1=Product("Laptop",50000)
print(p1.name,p1.price)


p2=Product("Headphones",500)
print(p2.name,p2.price)
```
