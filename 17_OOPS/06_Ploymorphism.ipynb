````python
## 🔷 Polymorphism in Python

**Polymorphism** means *“many forms”*. In Python, it allows the same function or method name to behave differently depending on the object or data type.

👉 In simple words:
**One interface → Multiple implementations**

---

# 🔹 Types of Polymorphism in Python

## 1. Method Overriding (Runtime Polymorphism)

When a **child class provides a different implementation** of a method that already exists in the parent class.

### ✅ Example:

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")


# Usage
a = Animal()
d = Dog()
c = Cat()

a.speak()
d.speak()
c.speak()
```

### 🔍 Output:

```
Animal makes sound
Dog barks
Cat meows
```

👉 Same method `speak()` behaves differently.

---

## 2. Method Overloading (Compile-time style)

Python does **not support true method overloading**, but we can achieve it using:

* Default arguments
* `*args`

### ✅ Example using default arguments:

```python
class Math:
    def add(self, a, b, c=0):
        return a + b + c

m = Math()

print(m.add(2, 3))       # 5
print(m.add(2, 3, 4))    # 9
```

👉 Same method works with different number of arguments.

---

## 3. Operator Overloading

Python allows operators (`+`, `-`, `*`) to work differently for objects using **magic methods**.

### ✅ Example:

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print(self.x, self.y)


p1 = Point(2, 3)
p2 = Point(4, 5)

p3 = p1 + p2
p3.display()
```

### 🔍 Output:

```
6 8
```

👉 `+` operator works for custom objects.

---

## 4. Duck Typing (Dynamic Polymorphism)

Python focuses on **behavior, not type**.

👉 “If it looks like a duck and behaves like a duck, it is a duck.”

### ✅ Example:

```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Airplane:
    def fly(self):
        print("Airplane flies")

def make_fly(obj):
    obj.fly()

b = Bird()
a = Airplane()

make_fly(b)
make_fly(a)
```

### 🔍 Output:

```
Bird can fly
Airplane flies
```

👉 No inheritance required — just same method name.

---

# 🔥 Summary Table

| Type                 | Description                                    |
| -------------------- | ---------------------------------------------- |
| Method Overriding    | Same method, different behavior in child class |
| Method Overloading   | Same method, different parameters              |
| Operator Overloading | Operators behave differently                   |
| Duck Typing          | Object behavior matters, not type              |

---

# 🚀 Real-Life Example

Think of a **remote control**:

* Same button → different TVs react differently
  👉 That’s polymorphism!

---

If you want, I can give:

* 💯 Interview questions on polymorphism
* 🧠 Tricky coding problems
* 📊 Diagrams for better understanding

Just tell me 👍
````

````python
# operator Overloading

# Python __add__() function is one of the magic methods in Python that returns a new object(third) i.e. the addition of the other two objects. It implements the addition operator "+" in Python.

# Note + -- any number of operands
#      __add__ :compulsary 2 operands 

# Operators behave differently

class Number:
    def __init__(self,val):
        self.val=val

    def __add__(self, other):
        return self.val+other.val
    

n1=Number(5)
n2=Number(10)
print(n1+n2)
````

````python
class Person:
    def __init__(self,age):
        self.age=age
     
    def __gt__(self, other):
        return self.age>other.age
    

p1=Person(25)
p2=Person(20)
print(p1>p2)
````

````python
class Multiply:
    def __init__(self,num):
        self.num=num

    def __mul__(self, other):
        return self.num*other.num
    

m1=Multiply(10)
m2=Multiply(20)
print(m1*m2)
````

````python
class Order:
    def calculate_total(self, amount):
        return amount

    def display(self, amount):
        print(f"Total Amount: ₹{self.calculate_total(amount)}")


class DineIn(Order):
    def calculate_total(self, amount):
        return amount + 50   # service charge

    def display(self, amount):
        print(f"Dine-In Order → Base: ₹{amount}, Service Charge: ₹50, Total: ₹{self.calculate_total(amount)}")


class TakeAway(Order):
    def calculate_total(self, amount):
        return amount + 20   # packaging charge

    def display(self, amount):
        print(f"TakeAway Order → Base: ₹{amount}, Packaging Charge: ₹20, Total: ₹{self.calculate_total(amount)}")


class Delivery(Order):
    def calculate_total(self, amount):
        return amount + 100  # delivery charge

    def display(self, amount):
        print(f"Delivery Order → Base: ₹{amount}, Delivery Charge: ₹100, Total: ₹{self.calculate_total(amount)}")


# Usage
orders = [DineIn(), TakeAway(), Delivery()]

for order in orders:
    order.display(500)
````

````python
class Loan:
    def calculate_interest(self, amount):
        return amount * 0.05

    def display(self, amount):
        print(f"Loan Interest: ₹{self.calculate_interest(amount)}")


class HomeLoan(Loan):
    def calculate_interest(self, amount):
        return amount * 0.07

    def display(self, amount):
        print(f"🏠 Home Loan → Amount: ₹{amount}, Interest: 7%, Total Interest: ₹{self.calculate_interest(amount)}")


class CarLoan(Loan):
    def calculate_interest(self, amount):
        return amount * 0.09

    def display(self, amount):
        print(f"🚗 Car Loan → Amount: ₹{amount}, Interest: 9%, Total Interest: ₹{self.calculate_interest(amount)}")


class PersonalLoan(Loan):
    def calculate_interest(self, amount):
        return amount * 0.12

    def display(self, amount):
        print(f"💳 Personal Loan → Amount: ₹{amount}, Interest: 12%, Total Interest: ₹{self.calculate_interest(amount)}")


# Usage
loans = [HomeLoan(), CarLoan(), PersonalLoan()]

for loan in loans:
    loan.display(100000)
````
