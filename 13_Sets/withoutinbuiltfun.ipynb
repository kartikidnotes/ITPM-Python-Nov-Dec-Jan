```python
# remove the duplicate

lst=[1,2,3,2,3,4,5,6]
result=[]

for i in lst:
    if i not in result:
        result.append(i)

print(result)
```

```python
# Union of two sets

a={1,2,3}
b={3,4,5}

res=[]

for i in a:
    res.append(i)

for j in b:
    if j not in res:
        res.append(j)

print(res)
```

```python
# intersection of two sets

a={1,2,3,4}
b={3,4,5,6}
res=[]

for i in a:
    if i in b:
        res.append(i)

print(res)
```

```python
# difference 

a={1,2,3,4}
b={2,3}
res=[]

for i in a:
    if i not in b:
        res.append(i)

print(res)
```
