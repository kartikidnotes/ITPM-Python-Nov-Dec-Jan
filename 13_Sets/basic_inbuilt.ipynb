```python
# find pair with given sum

arr=[1,2,3,4,5]
target=6

seen=set()

for num in arr:
    if target-num in seen:
        print(num,target-num)
    seen.add(num)
```

```python
# First Non Repeating elements

lst=[1,2,2,3,1,4]

for i in lst:
    if lst.count(i)==1:
        print(i)
        break
```

```python
# find duplicate

lst=[1,2,3,2,3,4,5,4,3]

seen=set()

for i in lst:
    if i in seen:
        print("Duplicate Element : ",i)
    else:
        seen.add(i)

print(seen)
```
