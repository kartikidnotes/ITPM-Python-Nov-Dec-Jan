```python
# create a dictionary

student={"id":1,"name":"Soham","marks":90}

print(student)
```

```python
# get value using key

student={"id":1,"name":"Soham","marks":90}

print(student.get("name"))
```

```python
# update dictonary 


student={"id":1,"name":"Soham","marks":90}

student["city"]="Mumbai"
student["course"]="Python "

print(student)
```

```python
# remove element:
d={"a":1,"b":2,"c":3,"d":2}

d.pop("b")

print(d)
```

```python
# get all keys,values

student={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

print(student.keys())
print(student.values())
print(student.items())
```

```python
# check whether key exists 

std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

if "name" in std:
    print("Key Exists")
else:
    print("Key Don't Exists")
```
