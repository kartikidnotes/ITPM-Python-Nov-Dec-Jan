```python
lst=[]

n=int(input("Enter number of elements of list : "))
for i in range(n):
    lst.append(int(input("Enter Elements : ")))

print("List is : ",lst)

original_lst = lst.copy()   # copy of list

original_lst.reverse()      # reverse the copy

if lst == original_lst:
    print("List is Palindrome")
else:
    print("List is not Palindrome")
```

```python
# check whether tuple contains duplicate elemenets or not

# t=(10,20,30,50,20,202,20,0,20,20)
t=(10,20,30)

duplicate=False

for i in t:
    count=0
    for j in t:
        if i==j:
            count=count+1
        
    if count>1:
        duplicate=True
if duplicate:
    print("Duplicate Elements are Present ")
else:
    print("Duplicate Elements are NOT Present ")
```
