```python
# even number 

def even_num(n):
    for i in range(2,n+1,2):
        print(i)

even_num(20)
```

```python
# with arg with parameter

def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact


n=int(input("Enter number :: "))
print("Factorial of num is :: ",factorial(n))
```

```python
def max_num(a,b):
    if a>b:
        return a
    return b

print(max_num(10,20))
```
