```python
# prove that conctructor is called automatically

class Demo:
    def __init__(self):
        print("Constructor is Called !!!!!")
    
#  created runtime object 
Demo()
```

```python
# Parametrized COnstructor

class Student:
    def __init__(self,rollno,name,courrse):
        self.rollno=rollno
        self.name=name
        self.course=courrse


s1=Student(1,"Ram","Python")
s2=Student(2,"Raj","AI/ML")

print(s1.rollno, s1.name, s1.course)
print(s2.rollno,s2.name,s2.course)
```

```python
class Student:
    def __init__(self,name,course):
        self.name=name
        self.course=course

    def display(self):
        print("======== Student Details ==========")
        print("NAme is :: ",self.name)
        print("Course is :: ",self.course)


#user input
n1=input("Enter Name :: ")
c1=input("Enter COurse Name :: ")
s1=Student(n1,c1)
s1.display()

n2=input("Enter Name :: ")
c2=input("Enter COurse Name :: ")
s2=Student(n2,c2)
s2.display()
```

```python
class Bank:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    def deposit(self,damt):
        self.balance+=damt
        print("Deposited AMount is :: ",damt)
        # print("Balance AMount is :: ",self.balance)

    def withdraw(self,wamt):
        self.balance-=wamt
        print("Withdraw AMount is :: ",wamt)
        # print("Balance AMount is :: ",self.balance)


name=input("Enter Account Holder Name :: ")
balance=float(input("Enter Balance :: "))


acc=Bank(name,balance)

damt=float(input("Enter Deposit AMount :: "))
acc.deposit(damt)

wamt=float(input("Enter Deposit AMount :: "))
acc.withdraw(wamt)
    
```

```python
# Product Details 

class Product:
    def __init__(self,name,price,quantity):
        self.name=name
        self.price=price
        self.quantity=quantity


    def total(self):
        return self.price*self.quantity
    

name=input("Enter Name :: ")
price=float(input("Enter Price :: "))
quantity=int(input("enter Quantity :: "))


p=Product(name,price,quantity)
print("Total Cost is :: ",p.total())
```
