```python
# Variable Length Argument 

def sum(*nums):
    s=0
    for i in nums:
        s+=i
    print(s)

sum(1,2,3,4,5)
sum(10,20,30)
sum(20,30,100)
```

```python
def show(*names):
    for i in names:
        print(i)

show("Ram","Raj")
show("Om","Ajay","Vijay","Raj")
```

```python
def max(*nums):
    max_value=nums[0]
    for i in nums:
        if i>max_value:
            max_value=i
    print(max_value)

max(1,9,0,22,8,0,0,2,3,7)
max(100,300,6,98,78,87777,90,0)
```

```python
def demo(*nums):
    d=0
    for i in nums:
        d+=i
    return d


n=int(input("enter count :: "))
lst=[]

for i in range(n):
    lst.append(int(input("Enter elements :: ")))

print(demo(*lst))
```

```python
# Dynamic Keyword Variable Length (**kwargs)

def details(**data):
    print(data)


details(name="Ram",salary=20000,designation="Developer")
details(name="Raj",marks=90,city="Pune")
```

```python
def show(**info):
    for k,v in info.items():
        print(k, " : ", v)


show(dept_name="IT",employee_count=4)
show(coursename="Python",version=3)
```

```python
def profile(**data):
    if "name" in data:
        print("Name is :: ",data["name"])


profile(name="Ram",city="Pune")
profile(name="Raj")
```

```python
# lambda Function -- Anonymous function

# add two number

add=lambda a,b:a+b
print(add(5,4))
```

```python
# square 

sqaure=lambda side:side*side
print(sqaure(5))
```

```python
# even odd 

even=lambda num:"Even number " if num%2==0 else "Odd Number"

num=int(input("Enter Number :: "))
print(even(num))
```

```python
# based on second element -- sorting 

lst=[(1,3),(2,1),(4,2)]
lst.sort(key=lambda n:n[1])
print(lst)
```

```python
# filter positive element

lst=[-2,3,-8,-6,5,8]
res=list(filter(lambda num:num>0,lst))
print(res)
```

```python
# Local Scope of variable 
# it is not a type of argument 

def demo():
    a=10
    print(a)

demo()

# if a variable is defined inside a function so the scope of the varibale is inside the function only 
# print(a)
```

```python
def demo():
    for i in range(1,6):
        print(i)

demo()

print("Hello")
# print(i)
```

```python
# global scope of variable

num=100

def demo():
    print(num*2)

# num=100
demo()
print(num)
```

```python
count=0

def update():
    global count
    count+=1


update()
print(count)
```

```python
balance=5000
print(balance)

def deposit():
    global balance
    balance+=2000

deposit()
print(balance)
```

```python

# Global and Local 

num=10

def test():
    # global num
    num=20
    print("Local Variable " , num)

test()
print("Global Variable " ,num)
```

```python
# Recursion 
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
# sum 

def sum(n):
    if n==0:
        return 0
    return n+sum(n-1)


print(sum(5))
```
