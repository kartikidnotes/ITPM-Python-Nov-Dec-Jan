```python
# count the length 

std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

count=0
for _ in std:
    count+=1

print(count)
```

```python
# CHECK KEY EXISTS

std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}
key="city"

found=False
for k in std:
    if k==key:
        found=True


print(found)
```

```python
# find maximum value 

d={"a":10,"b":50,"c":20}

values=[]
for i in d:
    values.append(d[i])

max_value=values[0]

for v in values:
    if v>max_value:
        max_value=v

print("Maximum Value is :: ",max_value)
```

```python
# merge two dictonary 

d1={"a":1,"b":2}
d2={"c":3,"d":4}

for i in d2:
    d1[i]=d2[i]

print(d1)
```

```python
# reverse the dictonary

# std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}
std={"a":1,"b":2,"c":3}
rev={}

for i in std:
    val=std[i]
    rev[val]=i


print(rev)
```

```python
# frequency count word 

s="india is my country"
freq={}

for ch in s:
    if ch in freq:
        freq[ch]=freq[ch]+1
    else:
        freq[ch]=1
        
print(freq)
```

```python
# remove key (pop())
std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

new_dict={}
key="name"

for i in std:
    if i!=key:
        new_dict[i]=std[i]

print(new_dict)
```

```python
# copy dictonary

std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

new_dict={}

for i in std:
    new_dict[i]=std[i]

print("New Dictonary is :: ",new_dict)
```

```python
# word frequency in sentence 

s="india is my country. india is in Asia Continent"

words=s.split()
freq={}

for i in words:
    freq[i]=freq.get(i,0)+1


print(freq)
```

```python
# Find the duplicate values

d={"a":10,"b":20,"c":10,"d":30,"e":20}

seen={}
duplicates=[]

for i in d:
    val=d[i]
    if val in seen:
        duplicates.append(val)
    else:
        seen[val]=1

print(duplicates)
```

```python
# key with maximum value

d={"a":10,"b":50,"c":40,"d":50}

max_key=""
max_val=0

for k in d:
    if d[k]>max_val:
        max_val=d[k]
        max_key=k

print(max_key ," : ",max_val )
```

```python
# group by word length

words=["hi","hello","dad","mom","cat","python"]

group={}

for i in words:
    l=len(i)
    if l not in group:
        group[l]=[]
    group[l].append(i)

print(group)
```

```python
# nested dictonary

students={
    "s1":{"name":"Ram","marks":90},
    "s2":{"name":"Raj","marks":80},
    "s3":{"name":"Jay","marks":70}
}

for s in students:
    print(students[s]["name"], students[s]["marks"])
```

```python
# handle the duplicates vlaues

d={"a":10,"b":20,"c":10}

dup={}

for i in d:
    val=d[i]
    if val not in dup:
        dup[val]=[]
    dup[val].append(i)

print(dup)
```

```python
# merge two dictonary 
a={"id":1,"name":"Ram"}
b={"course":"Python","duration":"2 months"}

for i in b:
    a[i]=b[i]

print(a)
```

```python
# Invert the dictonary -- swappp

d={"id":1,"name":"Ram","marks":90,"course":"python"}

inv={}

for i in d:
    value=d[i]
    inv[value]=i

print(inv)
```
