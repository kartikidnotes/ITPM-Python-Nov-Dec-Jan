```python
# Create a tuple  with ()

t=(10,20,30,40,50)

print(t)
```

```python
# create a tuple without ()

t=100,200,300

print(t)
```

```python
# single element in tuple

t=(200,)

print(t)
```

```python
# access the elements in tuple

t=(100,200,500,800,900)

print(t[0])
print(t[2])
# print(t[10])
```

```python
# tuple slicing

t=(10,20,30,40,50,60,70,80)

print(t[1:3])
print(t[:])
print(t[:4])
print(t[4:])
print(t[:-4])
print(t[:-5])
print(t[-5:-2])
# print(t[-2:-5])

```

```python
# tuple packing and unpacking

# tuple packing 
t=10,20,30,40,50
print(t)

# tuple unpacking
a,b,c=(10,20,30)
print(a)
print(b)
print(c)
```

```python
#tuple operation

# 1. concatenation
t1=(10,20,30)
t2=(100,200,300)

print(t1+t2)

# Repetation
t=(10,20)

print(t*5)
```

```python
# LENGTH OF TUPLE

t=(10,20,30,40,50,60,70,80)

print("Length of tuple is :: ",len(t))
```

```python
# Maximum and minimum element 

t=(10,5,9,99,87,67,56)

print("Minimum Element is :: ",min(t))
print("Maximum Element is :: ",max(t))

```

```python
# sum
t=(10,20,30,40,50,60,70)
print("Sum of  Element is :: ",sum(t))
```

```python
# Occurance of element Count
t=(10,20,30,10,20,30,40,50)

print("Count of tuple is :: ",t.count(10))
```

```python
# find index of particular element
t=(10,20,30,10,20,30,40,50)

print("Index of 40 is :: ",t.index(40))
print("Index of 30 is :: ",t.index(30))
```

```python
#  convert list to tuple

lst=[10,20,30,40,50,60,50]
t=tuple(lst)
print(t)
```

```python
# Sort tuple

t=(30,5,7,2,90,99)
print(sorted(t))
```

```python
# check element in tuple

t=(30,5,7,2,90,99)

print(30 in t)
print(9 in t)
```

```python
t=(30,5,7,2,90,99)

n=int(input("enter element to check :: "))

if n in t:
    print("Element Found ")
else:
    print("Element not found")
```

```python
# iterate the tuple

t=(10,9,0,7,8,9,55,8,0,87,76,65,676,764,89)

print(t)

for i in t:
    print(i)
```

```python
# Tuple reverse


t=(10,9,0,7,8,9,55,8,0,87,76,65,676,764,89)

rev=t[::-1]
print(t)
print(rev)
```

```python
# Print even number from tuple


t=(10,9,0,7,8,9,55,8,0,87,76,65,676,764,89)

for i in t:
    if i%2==0:
        print(i)
```

```python
# average of tuple


t=(10,20,30,40,50)

total=sum(t)

avg=total/len(t)

print("Average of tuple is ::  ",avg)
```

```python
# Student Record 

student=("Ram",90,60,80)

name=student[0]
total=student[1]+student[2]+student[3]

avg=total/3

print("Name is :: ",name)
print("Total MArks :: ",total)
print("Average is :: ",avg)
```
