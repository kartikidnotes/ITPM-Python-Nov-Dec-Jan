```python
# find the second highest eleemnt

t=(10,20,40,8,90,9,100)

max1=t[0]
max2=t[0]

for i in t:
    if i>max1:
        max2=max1
        max1=i
    elif i>max2 and i!=max1:
        max2=i

print("Second Higest Element is :: ",max2)
```

```python
# merge two tuples without duplicates

t1=(10,20,30,40)
t2=(30,40,50,60)

result=()

for i in t1:
    result=result+(i,)

for i in t2:
    found=False
    for j in result:
        if i==j:
            found=True
    if not found:
        result=result+(i,)

print(result)
```

```python
# find the frequency of each element in tuple

t=(10,20,50,60,20,20,20,10,10,20,30,60)

visited=()

for i in t:
    already=False
    for v in visited:
        if i==v:
            already=True
    if not already:
        count=0
        for j in t:
            if i==j:
                count=count+1
        print(i," == ",count)
        visited=visited+(i,)
```

```python
# split the tuples into even and odd tuples

t=(10,20,33,40,55,60)

even=()
odd=()

for i in t:
    if i%2==0:
        even=even+(i,)
    else:
        odd=odd+(i,)

print("Even tuple is :: ",even)
print("Odd Tuple is :: ",odd)
```

```python
# find the common elements

t1=(10,20,30,40,50)
t2=(30,40,50,60,70)

common=()

for i in t1:
    for j in t2:
        if i==j:
            found=False
            for k in common:
                if k==i:
                    found=True
            if not found:
                common=common+(i,)
print("Common Elements are :: ",common)
```
