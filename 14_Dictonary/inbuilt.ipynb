```python
# create a dictionary

student={"id":1,"name":"Soham","marks":90}

print(student)
```

```python
# get value using key

student={"id":1,"name":"Soham","marks":90}

print(student.get("name"))
```

```python
# update dictonary 


student={"id":1,"name":"Soham","marks":90}

student["city"]="Mumbai"
student["course"]="Python "

print(student)
```

```python
# remove element:
d={"a":1,"b":2,"c":3,"d":2}

d.pop("b")

print(d)
```

```python
# get all keys,values

student={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

print(student.keys())
print(student.values())
print(student.items())
```

```python
# check whether key exists 

std={'id': 1, 'name': 'Soham', 'marks': 90, 'city': 'Mumbai', 'course': 'Python '}

if "name" in std:
    print("Key Exists")
else:
    print("Key Don't Exists")
```

```python
# Mere two dictonary

d1={"id":1}
d2={"name":"Raj","course":"Python"}

d1.update(d2)
print(d1)
print(d2)
```

```python
# count the frequency of each character

s="india is my country"
freq={}

for ch in s:
    freq[ch]=freq.get(ch,0)+1

print(freq)
```

```python
# maximum value 

a={"python":3,"java":5,"os":10}

print(max(a.values()))
```

```python
# sort dict with keys

d={"c":2,"b":3,"d":10,"a":5}

for i in sorted(d):
    print(i,d[i])
```
