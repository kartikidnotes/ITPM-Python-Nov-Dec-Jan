```python
🔷 1. What is Abstraction in Python?

Abstraction = Hiding implementation details and showing only essential features.

Eg : 

You use a mobile phone without knowing internal circuits.
You call a function without knowing how it works internally.

🔹 Why Abstraction?
Reduces complexity
Improves code readability
Enhances security (hide internal logic)
Promotes reusability

🔹 How Python Achieves Abstraction?
1. Abstract Classes (using abc module)
2. Interfaces-like behavior (via abstract methods)
3. Encapsulation (partially supports abstraction)

🔹 Abstract Class Syntax
from abc import ABC, abstractmethod

class Shape(ABC):
    
    -- @ is called as decorator 
    @abstractmethod
    def area(self):
        pass


🔷 2. Key Rules
Cannot create object of abstract class
Must implement all abstract methods in child class
Used as a blueprint
```

```python
from abc import ABC,abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return 3.14*5.5*5.5

print(Circle().area())
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass


class Circle(Shape):
    def __init__(self,r):
        self.r=r

    def area(self):
        return 3.14*self.r*self.r
    

class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b

    def area(self):
        return self.l*self.b
       

r=float(input("Enter Radius :: "))
c=Circle(r)
print("Area Of cirlce is :: ",c.area())

l=int(input("Enter Length :: "))
b=int(input("Enter Breadth :: "))
rect=Rectangle(l,b)
print("Area Of Rectangle is :: ",rect.area())
```

```python
from abc import ABC,abstractmethod
# abc = package , ABC- ABSTRACT BASE CLASS , @ - decorator 

class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):
    def pay(self,amount):
        print("Paid RS. ",amount,"via UPI")

class Cash(Payment):
    def pay(self,amount):
        print("Paid RS. ",amount,"via Cash")

class Card(Payment):
    def pay(self,amount):
        print("Paid RS. ",amount,"via Card")


amount=int(input("Enter Amount :: "))
u=UPI()
u.pay(amount)

amount=int(input("Enter Amount :: "))
c=Cash()
c.pay(amount)

amount=int(input("Enter Amount :: "))
ca=Card()
ca.pay(amount)



```

```python
# The Universal Translator
# Goal: Create a system that "translates" a message into different formats without the user knowing the formatting logic.

# Task:

# Create an abstract class Translator with an abstract method display_message(text).

# Implement two subclasses:

# ShoutingTranslator: Converts the input text to ALL CAPS.

# WhisperingTranslator: Converts the input text to lowercase.

# User Input: Ask the user to type a sentence.

# Action: Create instances of both subclasses and pass the user's sentence into .display_message() for each to see the
#  different results.


from abc import ABC, abstractmethod

# 1. Create the Abstract Base Class
class Translator(ABC):
    @abstractmethod
    def display_message(self, text):
        pass

# 2. Implement the Subclasses
class ShoutingTranslator(Translator):
    def display_message(self, text):
        print(f"Shouting: {text.upper()}")

class WhisperingTranslator(Translator):
    def display_message(self, text):
        print(f"Whispering: {text.lower()}")

# 3. User Input
user_text = input("Enter a sentence to translate: ")

# 4. Action (Using Abstraction)
shouter = ShoutingTranslator()
whisperer = WhisperingTranslator()

shouter.display_message(user_text)
whisperer.display_message(user_text)
```

```python

# Assignment 4: Simple Appliance Controller
# Goal: Use abstraction to turn different home appliances on and off using a single command structure.

# Task:

# Create an abstract class Appliance with an abstract method turn_on().

# Implement two subclasses:

# Toaster: turn_on() should print: "Heating up the coils..."

# Lamp: turn_on() should print: "Illuminating the bulb..."

# User Input: Ask the user: "Which appliance would you like to use? (Toaster/Lamp): ".

# Action: Use an if/else statement to create the object chosen by the user, then call the .turn_on() method.




from abc import ABC, abstractmethod

# 1. Create the Abstract Base Class
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# 2. Implement the Subclasses
class Toaster(Appliance):
    def turn_on(self):
        print("Toaster: Heating up the coils... Watch your toast!")

class Lamp(Appliance):
    def turn_on(self):
        print("Lamp: Illuminating the bulb... The room is bright!")

# 3. User Input
choice = input("Which appliance would you like to use? (Toaster/Lamp): ").strip().capitalize()

# 4. Action (Selection and Execution)
if choice == "Toaster":
    device = Toaster()
elif choice == "Lamp":
    device = Lamp()
else:
    device = None
    print("Unknown appliance!")

# If a valid device was created, call the abstract method
if device:
    device.turn_on()
```

```python
from abc import ABC, abstractmethod

class Notification(ABC):
    @abstractmethod
    def send(self, msg):
        pass

class Email(Notification):
    def send(self, msg):
        print("Email sent:", msg)

class SMS(Notification):
    def send(self, msg):
        print("SMS sent:", msg)

text = input("Enter message: ")

Email().send(text)
SMS().send(text)
```

```python
from abc import ABC, abstractmethod

class Restaurant(ABC):
    @abstractmethod
    def show_menu(self):
        pass

class PizzaPlace(Restaurant):
    def show_menu(self):
        print("Pizza Menu: Margherita, Farmhouse")

class BurgerPlace(Restaurant):
    def show_menu(self):
        print("Burger Menu: Veg Burger, Cheese Burger")

choice = input("Choose Restaurant (Pizza/Burger): ").lower()

if choice == "pizza":
    obj = PizzaPlace()
elif choice == "burger":
    obj = BurgerPlace()
else:
    print("Invalid choice")
    obj = None

if obj:
    obj.show_menu()
```

```python
from abc import ABC, abstractmethod

class Course(ABC):
    @abstractmethod
    def start(self):
        pass

class PythonCourse(Course):
    def start(self):
        print("Python course started")

class DataScienceCourse(Course):
    def start(self):
        print("Data Science course started")

choice = input("Python/DataScience: ").lower()

if choice == "python":
    PythonCourse().start()
else:
    DataScienceCourse().start()
```

```python
from abc import ABC, abstractmethod

class Delivery(ABC):
    @abstractmethod
    def deliver(self):
        pass

class Standard(Delivery):
    def deliver(self):
        print("Standard delivery")

class Express(Delivery):
    def deliver(self):
        print("Express delivery")

choice = input("Standard/Express: ").lower()

if choice == "express":
    Express().deliver()
else:
    Standard().deliver()
```
